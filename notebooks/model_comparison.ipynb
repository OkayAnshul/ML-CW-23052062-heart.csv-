```python
import pandas as pd

results = pd.DataFrame({
    "Algorithm": [
        "Logistic Regression",
        "KNN",
        "Naive Bayes",
        "Decision Tree (Unpruned)",
        "Decision Tree (Pruned)"
    ],
    "Accuracy": [0.82, 0.87, 0.78, 0.99, 0.83]
})

results
```

```python
import matplotlib.pyplot as plt

plt.bar(results["Algorithm"], results["Accuracy"])
plt.xticks(rotation=45)
plt.ylabel("Accuracy")
plt.title("Model Comparison")
plt.show()
```
